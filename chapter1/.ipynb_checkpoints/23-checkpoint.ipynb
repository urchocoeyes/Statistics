```python
import random
import matplotlib.pyplot as plt

random.seed(42)
```

```python
# Define the sample space
sample_space = [1, 2, 3, 4, 5, 6]

# Define events A and B
A = {2, 4, 6}
B = {1, 2, 3, 4}
```

```python
# Theoretical probabilities
P_A = len(A) / len(sample_space)
P_B = len(B) / len(sample_space)
P_AB = len(A.intersection(B)) / len(sample_space)

print(f"P(A) = {P_A}")
print(f"P(B) = {P_B}")
print(f"P(AB) = {P_AB}")
```

```python
# Number of simulations
n_simulations = 100000

# Counters for each event
count_A = 0
count_B = 0
count_AB = 0

# Run simulations
for _ in range(n_simulations):
    roll = random.choice(sample_space)
    
    if roll in A:
        count_A += 1
    if roll in B:
        count_B += 1
    if roll in A and roll in B:
        count_AB += 1

# Calculate simulated probabilities
P_hat_A = count_A / n_simulations
P_hat_B = count_B / n_simulations
P_hat_AB = count_AB / n_simulations

print(f"Estimated P(A) = {P_hat_A}")
print(f"Estimated P(B) = {P_hat_B}")
print(f"Estimated P(AB) = {P_hat_AB}")
print(f"Estimated P(A)*P(B) = {P_hat_A * P_hat_B}")
```

```python
import random

random.seed(42)

# Define the sample space and events
sample_space = [1, 2, 3, 4, 5, 6]

# Independent Events
A_independent = {2, 4, 6}
B_independent = {1, 2, 3, 4}

# Theoretical probabilities for independent events
P_A_independent = len(A_independent) / len(sample_space) # 1 / 2
P_B_independent = len(B_independent) / len(sample_space) # 2 / 3
P_AB_independent = len(A_independent.intersection(B_independent)) / len(sample_space) # 2 / 6 = 1 / 3 yes, they are independent indeed

# Non-Independent Events
A_non_independent = {1, 2, 3}
B_non_independent = {2, 3, 4}

# Theoretical probabilities for non-independent events
P_A_non_independent = len(A_non_independent) / len(sample_space) # 1 / 2
P_B_non_independent = len(B_non_independent) / len(sample_space) # 1 / 2
P_AB_non_independent = len(A_non_independent.intersection(B_non_independent)) / len(sample_space) # 1 / 4 = 0.25 if they independent, but actually 2 / 6 = 1 / 3 is not equal to 0.25, that's why they are NOT INDEPENDENT

# Number of simulations
n_simulations = 100000

# Function to simulate events
def simulate_events(A, B):
    count_A = 0
    count_B = 0
    count_AB = 0
    
    for _ in range(n_simulations):
        roll = random.choice(sample_space)
        
        if roll in A:
            count_A += 1
        if roll in B:
            count_B += 1
        if roll in A and roll in B:
            count_AB += 1
            
    P_hat_A = count_A / n_simulations
    P_hat_B = count_B / n_simulations
    P_hat_AB = count_AB / n_simulations
    
    return P_hat_A, P_hat_B, P_hat_AB

# Simulate for independent events
P_hat_A_independent, P_hat_B_independent, P_hat_AB_independent = simulate_events(A_independent, B_independent)

# Simulate for non-independent events
P_hat_A_non_independent, P_hat_B_non_independent, P_hat_AB_non_independent = simulate_events(A_non_independent, B_non_independent)

# Print results
print("Independent Events:")
print(f"Theoretical P(A) = {P_A_independent}")
print(f"Theoretical P(B) = {P_B_independent}")
print(f"Theoretical P(AB) = {P_AB_independent}")
print(f"Estimated P(A) = {P_hat_A_independent}")
print(f"Estimated P(B) = {P_hat_B_independent}")
print(f"Estimated P(AB) = {P_hat_AB_independent}")
print(f"Estimated P(A)*P(B) = {P_hat_A_independent * P_hat_B_independent}\n")

print("Non-Independent Events:")
print(f"Theoretical P(A) = {P_A_non_independent}")
print(f"Theoretical P(B) = {P_B_non_independent}")
print(f"Theoretical P(AB) = {P_AB_non_independent}")
print(f"Estimated P(A) = {P_hat_A_non_independent}")
print(f"Estimated P(B) = {P_hat_B_non_independent}")
print(f"Estimated P(AB) = {P_hat_AB_non_independent}")
print(f"Estimated P(A)*P(B) = {P_hat_A_non_independent * P_hat_B_non_independent}")
```
