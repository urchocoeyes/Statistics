```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate coin flips for p = 0.3
p_1 = 0.3
n = 1000
flips_1 = np.random.binomial(1, p_1, n) # This function simulates the process of flipping a biased coin n times.  
proportion_heads_1 = np.cumsum(flips_1) / (np.arange(1, n + 1)) # This operation divides the cumulative sum of heads by the corresponding number of flips at each point.

# Simulate coin flips for p = 0.03
p_2 = 0.03
flips_2 = np.random.binomial(1, p_2, n)
proportion_heads_2 = np.cumsum(flips_2) / (np.arange(1, n + 1))

# Plot proportion of heads as a function of n for both p values
plt.figure(figsize=(10, 6))
plt.plot(proportion_heads_1, label='p = 0.3')
plt.plot(proportion_heads_2, label='p = 0.03')
plt.axhline(y=p_1, color='blue', linestyle='--', label='Expected p = 0.3')
plt.axhline(y=p_2, color='orange', linestyle='--', label='Expected p = 0.03')
plt.xlabel('Number of Coin Flips')
plt.ylabel('Proportion of Heads')
plt.title('Proportion of Heads as a Function of Number of Coin Flips')
plt.legend()
plt.grid(True)
plt.show()
```
